```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
data = pd.read_csv('./DATA/extra_Breast_Cancer.csv')
```

```python
data
```

```python
data.describe()
```

```python
y =data['Survival Months'] 
x1 = data['Tumor Size']
```

```python
plt.scatter(x1,y)
plt.xlabel('Tumor Size', fontsize = 20)
plt.ylabel('Survival Months', fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x1) 
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = -0.0943*x1 + 74.1721
fig = plt.plot(x1,yhat, lw=4, c='orange', label ='regression line')
plt.xlabel('Tumor Size', fontsize = 20)
plt.ylabel('Survival Months', fontsize = 20)
plt.show()
```
